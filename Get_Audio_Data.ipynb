## Import Libraries, and reading .wav file.

```python
import numpy as np
from scipy.io import wavfile
sample_rate, data_audio = wavfile.read('audio_sourc.wav')
```

##converting data into 32-bit integers.

```python
data = np.int32(data_audio[:,0])
```

###Printing some data.. You can skip this cell.

```python
for d in data[5000:5020]:
  print(hex(d))
```

## Saveing the 32-bit words into a file named 'Audio_data.txt'

```python
out = open('Audio_data.txt', 'w')

for i in range(len(data)):
  out.write('\t.word\t'+hex(data[len(data)-1-i])+'\n')

out.close()
```
